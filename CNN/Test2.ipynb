```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Activation, MaxPooling2D, Dense, Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
import keras
import random
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dropout
```

```python
x_train = np.loadtxt(r"F:\New folder\Image Classification CNN Keras Dataset-20240817T045957Z-001\Image Classification CNN Keras Dataset\input.csv",delimiter=',')
x_test = np.loadtxt(r"F:\New folder\Image Classification CNN Keras Dataset-20240817T045957Z-001\Image Classification CNN Keras Dataset\input_test.csv",delimiter=',')

y_train = np.loadtxt(r"F:\New folder\Image Classification CNN Keras Dataset-20240817T045957Z-001\Image Classification CNN Keras Dataset\labels.csv",delimiter=',')
y_test = np.loadtxt(r"F:\New folder\Image Classification CNN Keras Dataset-20240817T045957Z-001\Image Classification CNN Keras Dataset\labels_test.csv",delimiter=',')
```

```python
x_train = x_train.reshape(len(x_train),100,100,3)
x_test = x_test.reshape(len(x_test),100, 100, 3)

y_train = y_train.reshape(len(y_train),1)
y_test = y_test.reshape(len(y_test),1)

x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

y_train = y_train.astype('float32') / 255.0
y_test = y_test.astype('float32') / 255.0
```

```python
print("Shape of X_trained",x_train.shape)
print("Shape of X_tested",x_test.shape)
print("Shape of Y_trained",y_train.shape)
print("Shape of Y_tested",y_test.shape)
```

# Model Training

```python
model = Sequential([
    Conv2D(32, (3,3),activation = 'relu',input_shape=(100, 100, 3)),
    MaxPooling2D((2,2)),
    
    Conv2D(32, (3,3),activation = 'relu'),
    MaxPooling2D((2,2)),

    Flatten(),
    Dense(64, activation= 'relu'),
    Dense(1, activation= 'sigmoid')
])
```

# Model Training

```python
# opt = keras.optimizers.SGD(learning_rate=0.001)
model.compile(loss= 'binary_crossentropy',optimizer= 'adam', metrics= ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 15, batch_size=64) #run only this line 
# again and again to resume the training
```
